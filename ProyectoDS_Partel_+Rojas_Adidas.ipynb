```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importar Data de Archivo csv y mostrar informacion

```python
# Leer el archivo Excel
df = pd.read_excel('DF_Adidas.xlsx')

# Mostrar las primeras filas del dataframe
df.head()
```

```python
# Indicar numero de filas y de columnas de Data
df.shape
```

```python
# Indicar estadísticas descriptivas para cada columna numérica del DataFrame.
df.describe ()
```

```python
# Calcular valores nulos en el DataFrame
df.isnull().sum()
```

```python
# Remplaza valores nulos con Nan
df.replace(0, np.nan)
```

##Graficos

```python
# 1er Grafico - Distribucion de ventas por %

# Hipotesis: Evidenciar si el menor Porcentaje de ventas registradas corresponde a la categoria de futbol.
## Para esto realizamos un grafico de torta que mostrara el porcentaje de las ventas para cada una de las categorias informadas en los datos.

category_sales = df.groupby('Category')['Total_Sales'].sum()

plt.figure(figsize=(8, 6))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', colors=['gainsboro', 'cornflowerblue', 'mediumslateblue', 'lightskyblue'])
plt.title('Distribucion de Ventas Totales por Categoria')
plt.axis('equal')
plt.show()

#Respuesta: Nuestra hispotesis fue falsa, ya que a partir del grafico se evidencia que la categoria que representa el menor porcentaje de ventas es la de "Entrenar"
```

```python
# 2do Grafico - Numero de Ventas por forma

#Hipotesis: Hay mayor cantidad de ventas en tienda que en el resto de los metodos de venta informados.
##Para Verificcar si esta hipotesis es verdadera o falsa se creo un grafico de barras que mostrara las unidades vendidas por metodo de venta.

platform_counts = df['Sales_Method'].value_counts()
platform_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Metodo de Venta')
plt.ylabel('Cantidad de Ventas')
plt.title('Numero de Ventas por Metodo')
plt.xticks(rotation=45)
plt.show()

#Respuesta a Hispotesis: Nuestra hipotesis es falsa, ya que a partir del grafico se puede evidenciar que el mayor numero de ventas se realizo por via "Online".
```

```python
# 3er Grafico - Ganancias por Unidades vendidas por Genero

# Hipótesis: Hipótesis: Los hombres tienden a comprar más unidades de productos de Adidas en comparación con las mujeres, niños y niñas, incluso cuando las ventas totales son similares.
## Para verificar esta hipótesis, vamos a analizar el gráfico de dispersión y observar las tendencias de los datos para cada género.

#Filtrar por género
women_data = df[df['Gender'] == 'Women']
men_data = df[df['Gender'] == 'Men']
boy_data = df[df['Gender'] == 'Boy']
girl_data = df [df['Gender'] == 'Girl']

# Diagrama de dispersión para cada género
plt.figure(figsize=(10, 6), facecolor='lightgray')
plt.scatter(women_data['Total_Sales'], women_data['Units_Sold'], color='m', label='Women', alpha=0.8, edgecolor='gray')
plt.scatter(men_data['Total_Sales'], men_data['Units_Sold'], color='skyblue', label='Man', alpha=0.8, edgecolor='gray')
plt.scatter(boy_data['Total_Sales'], boy_data['Units_Sold'], color='lightgreen', label='Boy', alpha=0.8, edgecolor='gray')
plt.scatter(girl_data['Total_Sales'], girl_data['Units_Sold'], color='pink', label='Girl', alpha=0.8, edgecolor='gray')

plt.xlabel('Ventas Totales')
plt.ylabel('Unidades Vendidas')
plt.title('Ventas por Genero')
plt.legend()
plt.grid(True)
plt.show()

#Respuesta a Hipotesis: Los puntos de los datos para las hombres están más altos (es decir, más unidades vendidas) a lo largo del eje Total_Sales, esto sugiere que los hombres compran más unidades en promedio de productos de adidas que el resto de la poblacion evaluada (mujeres, ninos y ninas).
```

```python
# 4to Grafico - Ventas Totales por Estado

# Hipotesis: Determinar si en Nueva York representa la mayor cantidad de ventas de Adidas.
## Se crea grafico a partir de las ventas en tiendas por estado para determinar hipotesis.

from matplotlib.ticker import FuncFormatter

# Agrupar los datos por 'State' y sumar las ventas totales
State_sales = df.groupby('State')['Total_Sales'].sum().reset_index()

# Ordenar los datos por ventas totales en orden descendente
State_sales = State_sales.sort_values(by='Total_Sales', ascending=False)

# Función para formatear los valores como moneda
def currency(x, pos):
    return "${:,.0f}".format(x)

# Crear el gráfico de barras
plt.figure(figsize=(10, 8), facecolor='lightgray')
sns.barplot(x='Total_Sales', y='State', data=State_sales)

# Personalizar el gráfico
plt.xlabel('Ventas Totales')
plt.ylabel('Estado')
plt.title('Ventas Totales por Estado')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Aplicar el formateo de moneda al eje x
formatter = FuncFormatter(currency)
plt.gca().xaxis.set_major_formatter(formatter)

# Mostrar el gráfico
plt.show()

#Respuesta: A partir del grafico podemos certficar que la mayor cantidad de ventas se centra en el estado de Nueva York
```

```python
# 5to - Gráfico de líneas

#Hipotesis: Determinar si las ventas por region se inclinan a ventas de $600000
## Para este creamos un grafico de lineas para ver el comportamiento de las ventas por region

# Función para formatear los valores como moneda
def currency(x, pos):
    return "${:,.0f}".format(x)

plt.figure(figsize=(20, 10))
sns.lineplot(x='Total_Sales', y='Units_Sold', hue='Region', data=df, palette='Set2', ci=None)
plt.xlabel('Ventas Totales')
plt.ylabel('Unidades Vendidas')
plt.title('Ventas por Region')
plt.legend(title='Regiones')

# Aplicar el formateo de moneda al eje x
formatter = FuncFormatter(currency)
plt.gca().xaxis.set_major_formatter(formatter)

plt.show()

##Respuesta de Hipotesos: A partir del grafico podemos evidenciar que la mayor cantidad de ventas representa ventas menor
```
